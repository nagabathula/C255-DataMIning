```python
#dropping Route as this is not useful
airline_ticket_prices = airline_ticket_prices.drop(['Route'],axis=1)
```

```python
#Airlines value is changed to 0,1,2,3

airlines = list(airline_ticket_prices['Airline'].value_counts().index)
airlines_map = {k: v for v, k in enumerate(airlines)}
airline_ticket_prices['Airline'] = airline_ticket_prices['Airline'].map(airlines_map)
```

```python
#convert Date of JOurney to month , day and year columns

data = airline_ticket_prices['Date_of_Journey'].str.split('/',expand=True)

airline_ticket_prices['day']=data[0]
airline_ticket_prices['month']=data[1]
airline_ticket_prices['year']=data[2]
```

```python
#converinting the Dept and Arrival time into minutes
airline_ticket_prices['Dep_Time'] = pd.to_datetime(airline_ticket_prices['Dep_Time'])
airline_ticket_prices['Arrival_Time'] = pd.to_datetime(airline_ticket_prices['Arrival_Time'])

airline_ticket_prices['Dep_Time'] = airline_ticket_prices['Dep_Time'].apply(lambda x: x.hour * 60 + x.minute)
airline_ticket_prices['Arrival_Time'] = airline_ticket_prices['Arrival_Time'].apply(lambda x: x.hour * 60 + x.minute)

```

```python
#Convert the Duration into minutes 
airline_ticket_prices['H'] = airline_ticket_prices['Duration'].str.extract(r'(\d*(?=h))')
airline_ticket_prices['M'] = airline_ticket_prices['Duration'].str.extract(r'(\d*(?=m))')
airline_ticket_prices[['H', 'M']] = airline_ticket_prices[['H', 'M']].fillna(0)
airline_ticket_prices[['H', 'M']] = airline_ticket_prices[['H', 'M']].astype(int)
airline_ticket_prices['Duration'] = airline_ticket_prices['H'] * 60 + airline_ticket_prices['M']
```

```python
# replacing Total stops column into numerical

airline_ticket_prices['Total_Stops'] = airline_ticket_prices['Total_Stops'].replace({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4})

```

```python
# derive Included Bagge or Included Meals features from Additional Info 
airline_ticket_prices['Included_Baggage'] = airline_ticket_prices['Additional_Info'].apply(lambda x: 0 if x == 'No check-in baggage included' else 1)
airline_ticket_prices['Included_Meal'] = airline_ticket_prices['Additional_Info'].apply(lambda x: 0 if x == 'In-flight meal not included' else 1)
```

```python
# Source and Destination encoding (One-hot)
airline_ticket_prices = pd.concat([airline_ticket_prices, pd.get_dummies(airline_ticket_prices['Source'], prefix='S')], axis=1)
airline_ticket_prices = pd.concat([airline_ticket_prices, pd.get_dummies(airline_ticket_prices['Destination'], prefix='D')], axis=1)
    
# Fill in the single remaining missing value (Total Stops) with 0
```

```python
#Drop unused columns 
airline_ticket_prices = airline_ticket_prices.drop(['Source', 'Destination','Additional_Info','Date_of_Journey'], axis=1)

airline_ticket_prices = airline_ticket_prices.fillna(0)
```
